```python
all_data = np.load("data_all.npy file Path", allow_pickle =True)

```

```python
def label(data):
    lbll = [] #labels
    x = [] # images data
    #ft = [] # HOG features
    for i in range(0, len(data)):
        y = data[i][1]
        x_ = data[i][0]
        #ft_ = data[i][2]
        lbll.append(y)
        x.append(x_)
        #ft.append(ft_)
    return (x, lbll)

x, y = label(all_data)

```

```python
main_path = "/content/drive/MyDrive/Bitirme-Tasarım"
for name in CATAGORY:
  exist = os.path.exists(os.path.join(main_path, name))
  if not exist:
    os.makedirs(os.path.join(main_path,name))
  items = item["{}".format(name)]
  for cnt in range(len(items)):
    img_array = items[cnt][0]
    image = im.fromarray((img_array * 255).astype(np.uint8))
    label = items[cnt][1]
    saving_path = main_path + "/" + label
    image.save("{}/{}/{}{}.png".format(main_path,label,name,cnt))
```
